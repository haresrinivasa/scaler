```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = [2, 1] # weight w1 = 2; w2 = 1
w0 = 1 # bias

def plot_vector_and_line(w, w0):
    line_x = np.linspace(-5, 5, 100)
    line_y = (-w[0]*line_x - w0)/w[1] 
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.plot(line_x, line_y)

    ax.set_xlim(-5, 5)
    ax.set_ylim(-5, 5)
    
    arrowprops = {
        "arrowstyle": "->",
        "linewidth": 2,
        "color": "red",
    }
    ax.annotate("", w, [0, 0], arrowprops=arrowprops)
    
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.yaxis.tick_left()
    ax.xaxis.tick_bottom()
    ax.set_aspect("equal")

    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')

plot_vector_and_line(w=w, w0=w0)
```

```python
def distance_with_sign(x, w, w0):
    return (np.dot(x, w)+w0)/np.linalg.norm(w)
```

```python
distance_with_sign(x=[0.5, -0.1], w=w, w0=w0)
```

```python
def positive_or_negative(x, w, w0):
    return np.sign(np.dot(x, w) + w0)
```

```python
positive_or_negative(x=[0.5, 0], w=w, w0=w0)
```
